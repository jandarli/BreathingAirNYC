```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('asthmaset2.csv')

df
```

```python
plt.scatter(df.ZIPCode, df.Discharges, s=50)
plt.show()

loc = {0:"Manhattan", 1:"Bronx", 2:"Brooklyn", 3:"Queens", -1:"Outlier"}

df['subset'] = np.select([df.ZIPCode < 10280, (df.ZIPCode < 10805), df.ZIPCode < 11251, df.ZIPCode < 11697],
                         [0, 1, 2, 3], -1)
for color, label in zip('bgrym', [0, 1, 2, 3, -1]):
    subset = df[df.subset == label]
    plt.scatter(subset.ZIPCode, subset.Discharges, s=80, c=color, label= loc[label])
    
plt.legend(loc=1,prop={'size':9})

plt.title("Asthma Dischage Locations")
plt.show()

#manhattan 10001 -10280,bronx 10501 -10805,brooklyn 11201-11251, queens 11004 -11697
```
